```python
motifDB_ms1, motifDB_ms2 = load_motifDB("../../MS2LDA/MotifDB/motifDB.json")
```

```python
from massql4motifs import msql_engine
```

```python
massql_query_1 = "QUERY scaninfo(MS2DATA) WHERE MS2PROD=178.03"
```

```python
query_result_1 = msql_engine.process_query(massql_query_1, ms1_df=motifDB_ms1, ms2_df=motifDB_ms2)
```

```python
query_result_1
```

This query uses the normal massql syntax and returns all motifs containing fragment 178.03

```python
motifDB2motifs(motifDB_ms2, query_result_1)
```

By combining the motifDB with the query result we can extract all hits as spectrum objects 

Our modified version of massql can also search for exact matches in the metadata. Therefore we need to use the keyword METAFILTER:metadata=match. <br>
Let's say we want all results but only from measured with orbitraps

```python
massql_query_2 = "QUERY scaninfo(MS2DATA) WHERE MS2PROD=178.03 METAFILTER:analysis_massspectrometer=Orbitrap"
```

```python
query_result_2 = msql_engine.process_query(massql_query_2, ms1_df=motifDB_ms1, ms2_df=motifDB_ms2)
```

```python
query_result_2
```

Now we get all motifs with a fragment at 178.03 and measured by orbitrap. The METAFILTERs can also be combined.

```python
massql_query_3 = "QUERY scaninfo(MS2DATA) WHERE MS2PROD=102"
```

```python
query_result_3 = msql_engine.process_query(massql_query_3, ms1_df=motifDB_ms1, ms2_df=motifDB_ms2)
```

```python
query_result_3
```
