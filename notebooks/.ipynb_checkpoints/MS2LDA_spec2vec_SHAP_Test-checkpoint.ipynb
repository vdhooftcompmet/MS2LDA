### Motif analysis

```python
from Spec2Vec_annotation import load_model_and_data
from Spec2Vec_annotation import calc_similarity
from Spec2Vec_annotation import retrieve_top_hits
```

```python
s2v_model, embeddings_smiles_DB = load_model_and_data()
```

```python
smiles = embeddings_smiles_DB.smiles.to_numpy()
embeddings_DB = embeddings_smiles_DB.embeddings.to_list()
spectra_DB = embeddings_smiles_DB.spectra.to_list()
```

```python
s2v_scores = calc_similarity(s2v_model, motif_spectra, embeddings_DB)
```

```python
top_scores, top_smiles, top_spectra = retrieve_top_hits(s2v_scores, mn, smiles, spectra_DB, 10)
```

```python
top_scores
```

```python
from rdkit.Chem.inchi import MolToInchi
from rdkit.Chem.inchi import MolFromInchi
```

```python
unique_top_inchis = []
unique_top_spectra = []
unique_top_scores = []
for top_smi, top_spec, top_score in zip(top_smiles, top_spectra, top_scores):
    top_mol = Chem.MolFromSmiles(top_smi)
    top_inchi = MolToInchi(top_mol)
    if top_inchi not in unique_top_inchis:
        unique_top_inchis.append(top_inchi)
        unique_top_spectra.append(top_spec)
        unique_top_scores.append(top_score)
```

```python
Draw.MolsToGridImage([Chem.MolFromInchi(smi) for smi in unique_top_inchis], molsPerRow=5)
```

```python
from rdkit.Chem import rdFMCS
```

```python
mcs = rdFMCS.FindMCS([Chem.MolFromSmiles(smi) for smi in top_smiles], matchValences=True)
```

```python
Draw.MolToImage(Chem.MolFromSmarts(mcs.smartsString))
```

### Masking

```python
from Frag_and_loss_importance import mask_fragments, mask_losses
from Frag_and_loss_importance import masked_motif_similarity
from Frag_and_loss_importance import cluster_by_similarity
```

```python
motif = motif_spectra[mn]
```

```python
masked_fragment_motif = mask_fragments(motif, 1.00)
masked_loss_motif = mask_losses(motif, 1.00)
masked_motif = masked_fragment_motif + masked_loss_motif
```

```python
similarity_per_candidates = masked_motif_similarity(unique_top_spectra, masked_motif)
```

```python
clusters, Z, similarity_matrix = cluster_by_similarity(similarity_per_candidates)
```

```python
clusters
```

```python
import matplotlib.pyplot as plt
```

```python
print(len(unique_top_spectra))
plt.plot(np.array(similarity_per_candidates).T)
```

```python
from scipy.cluster.hierarchy import dendrogram
```

```python
plt.figure(figsize=(8, 6))
dendrogram(Z)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()
```

### cluster analysis

```python
grouped_smiles_1 = []
grouped_smiles_2 = []
for cluster, top_smi in zip(clusters, unique_top_inchis):
    if cluster == 1:
        grouped_smiles_1.append(top_smi)
    else:
        grouped_smiles_2.append(top_smi)
```

```python
Draw.MolsToGridImage([Chem.MolFromInchi(smi) for smi in grouped_smiles_1], molsPerRow=5)
```

```python
Draw.MolsToGridImage([Chem.MolFromInchi(smi) for smi in grouped_smiles_2], molsPerRow=5)
```

```python
L = 3
plt.plot(np.array(similarity_per_candidates[L]).T)
plt.hlines(unique_top_scores[L],0,20, color="red")
plt.show()

motif_frag_loss = list(motif_spectra[mn].peaks.mz) + list(motif_spectra[mn].losses.mz)
for mask_idx, similarity in enumerate(similarity_per_candidates[L]):
    if similarity < unique_top_scores[L]:
        print(str(motif_frag_loss[mask_idx]) + "  -->  " + str(similarity))
```
