```python
import pyLDAvis.gensim
pyLDAvis.enable_notebook()

import numpy as np
from CDK_pywrapper import FPType

from rdkit.Chem import Draw
from rdkit import Chem
from itertools import combinations
```

```python
import sys
sys.path.append('../programming_scripts/')

from MS2LDA_core import run_lda
from MS2LDA_core import predict_with_lda
from SMART_annotation import annotate_motifs
```

```python
spectra_path = "/Users/rosinatorres/Documents/PhD/Project/Data/PDE5_standards_annotated_pos_unique.mgf"
```

```python
lda_model, corpus, id2word = run_lda(spectra_path=spectra_path, num_motifs=6)
```

```python
pyLDAvis.gensim.prepare(lda_model, corpus, dictionary=lda_model.id2word)
```

```python
smiles_per_motifs, predicted_motifs, predicted_motifs_distribution, spectra_per_motifs = predict_with_lda(lda_model, spectra_path, id2word)
```

```python
import matplotlib.pyplot as plt

from MS2LDA_vis import extract_motifs
from MS2LDA_vis import sep_frags_losses
from MS2LDA_vis import frags_losses_2num
from MS2LDA_vis import ext_spec
from MS2LDA_vis import matching_motifs_spectra
from MS2LDA_vis import plot_spectra_with_motifs

extract_mot=extract_motifs(lda_model)
print(extract_mot)
```

```python
frags, losses = sep_frags_losses(extract_mot) #We separate the fragments and losses
```

```python
frags_2num=frags_losses_2num(frags)
losses_2num=frags_losses_2num(losses)

print(frags_2num)
print(losses_2num)
```

```python
#Here we need to extract the peaks and intensities from the list of objects spectra
select_motif=spectra_per_motifs[0] #Select the motif you want to plot
ext_peaks,ext_int=ext_spec(select_motif)
```

```python
#This finds all the motifs that are in the spectra, pairing motif with spectra, doing it for the length of the motif that was assign to
matches_found_frags=matching_motifs_spectra(select_motif,ext_peaks, ext_int, frags_2num)
matches_found_losses=matching_motifs_spectra(select_motif,ext_peaks, ext_int, losses_2num)

print(matches_found_frags)
```

```python
#We plot the frags and losses in the motif - and the spectras assigned to the motif
plot_spectra_with_motifs(ext_peaks, ext_int, matches_found_frags,matches_found_losses)
```
